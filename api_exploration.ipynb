```python
# Import dependencies
import requests
import json
import pandas as pd
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

FDA_KEY = os.getenv("FDA_API_KEY")

# fda_url = https://api.fda.gov/
```

```python
url = 'https://api.fda.gov/food/enforcement.json?search=state:New+York+AND+city:New+York&limit=5&api_key=' + FDA_KEY

res = requests.get(url)

res_json = res.json()

print(json.dumps(res_json, indent=4))
```
